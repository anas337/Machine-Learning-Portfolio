## Defining a Performance Metric
For this project, the target variable (prices) we want to predict is numeric and continuous so the problem we are trying to solve for this project is a **Regression Problem**. 

The output of any regression model already trained on this dataset will be house prices which is vector containing the the predicted price of each house. One of the methods used to measure or to quantify the performance of a regression model is the [*coefficient of determination*](http://stattrek.com/statistics/dictionary.aspx?definition=coefficient_of_determination), R<sup>2</sup>.

The coefficient of determination for a model is a useful statistic in regression analysis, as it often describes how "good" that model is at making predictions. 

The values for R<sup>2</sup> range from 0 to 1, which captures the percentage of squared correlation between the predicted and actual values of the **target variable**. A model with an R<sup>2</sup> of 0 is no better than a model that always predicts the *mean* of the target variable, whereas a model with an R<sup>2</sup> of 1 perfectly predicts the target variable. Any value between 0 and 1 indicates what percentage of the target variable, using this model, can be explained by the **features**. 

**A model can be given a negative R<sup>2</sup> as well, which indicates that the model is **arbitrarily worse** than one that always predicts the mean of the target variable.**

In the code cell below we will define the `performance_metric` function: 

- We will Use `r2_score` from `sklearn.metrics` to perform a performance calculation between `y_true` and `y_predict` and we will Assign the performance score to the `score` variable.

```python
# Import 'r2_score'
from sklearn.metrics import r2_score

# Defining the performance metric function 
def performance_metric(y_true, y_predict):
    """ This function Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    
    # Calculate the performance score between 'y_true' and 'y_predict'
    score = r2_score(y_true, y_predict)
    
    # Return the score
    return score
```

## Learning Curves:

**Definition: Learning Curves in Machine Learning** are Graphs that compares the performance of a model (Y-axis) on training and testing data over a varying number of training instances(X-axis). We should generally see **performance** improve as the number of training points increases. When we separate training and testing sets and graph them individually. We can get an idea of how well the model can generalize to new data. 

**Objective:**Learning Curves allow us to verify when a model has learnt as much as it can about the data. When it occurs The performances on the training and testing sets reach a plateau. There is a consistent gap between the two error rates The key is to find the sweet spot that minimizes bias and variance by finding the right level of model complexity. Of course with more data any model can improve, and different models may be optimal.


**Data Splitting:** train and test mean scores are generated each time before an amount of data is added. These scores are the mean of vector_scores. The lenght of this vector depends on the number of iteration we want to use  i.e how many time we want to train and test the model on the same amount (training size) of data before generating the mean scores related to these Data size. In order to enable that the data should be shuffled and splitted differently. Shuffle_split and KFold methods from cross validation are two common ways to shuffle and split and the data.

* **KFold VS Shuffle_split :**
    
    - By using KFold method will divide the data set into prespecified number of folds (subsets) of the full dataset, and every sample must be in one and only one fold. 
    
    - The ShuffleSplit method will randomly sample the entire dataset during each iteration to generate a training set and a test set. The test_size and train_size parameters control how large the test and training sets should be for each iteration. Since we are sampling from the entire dataset during each iteration, values selected during one iteration, could be selected again during another iteration.

**Performance Measurment:** The models performances could be mesured using error rates or scores. In this case we decide to mesure performance using RÂ² defined earlier. This score will be included in the learing curve method.  

Since we plan to use the entire dataset to generate learning curves we will use the ShuffleSplit method:

* It works iteratively, 
* More dynamic: the number of iterations is independent from the number of training subsets.

In the following code cell we will define the learning curve function using the shuffle split method and we will apply it to models mentioned earlier.

```python
from sklearn.cross_validation import ShuffleSplit
import sklearn.learning_curve as curves

def ModelLearning(X, y):
    """ Calculates the performance of several models with varying sizes of training data.
        The learning and testing scores for each model are then plotted. """
    # X: features 
    # y: target
    
    # Create 10 cross-validation sets for training and testing
    cv = ShuffleSplit(X.shape[0], n_iter = 10, test_size = 0.2, random_state = 337)

    # Generate the training set sizes increasing by 50
    train_sizes = np.rint(np.linspace(1, X.shape[0]*0.8 - 1, 9)).astype(int)

    # Create the figure window
    fig = plt.figure(figsize=(10,7))

    # Create three different models based on max_depth
    for k, depth in enumerate([1,3,6,10]):
        
        # Create a Decision tree regressor at max_depth = depth
        regressor = DecisionTreeRegressor(max_depth = depth)

        # Calculate the training and testing scores
        sizes, train_scores, test_scores = curves.learning_curve(regressor, X, y, \
            cv = cv, train_sizes = train_sizes, scoring = 'r2')
        
        # Find the mean and standard deviation for smoothing
        train_std = np.std(train_scores, axis = 1)
        train_mean = np.mean(train_scores, axis = 1)
        test_std = np.std(test_scores, axis = 1)
        test_mean = np.mean(test_scores, axis = 1)

        # Subplot the learning curve 
        ax = fig.add_subplot(2, 2, k+1)
        ax.plot(sizes, train_mean, 'o-', color = 'r', label = 'Training Score')
        ax.plot(sizes, test_mean, 'o-', color = 'b', label = 'Testing Score')
        ax.fill_between(sizes, train_mean - train_std, \
            train_mean + train_std, alpha = 0.15, color = 'r')
        ax.fill_between(sizes, test_mean - test_std, \
            test_mean + test_std, alpha = 0.15, color = 'b')
        
        # Labels
        ax.set_title('max_depth = %s'%(depth))
        ax.set_xlabel('Number of Training Points')
        ax.set_ylabel('Score')
        ax.set_xlim([0, X.shape[0]*0.8])
        ax.set_ylim([-0.05, 1.05])
    
    # Visual Esthetics
    ax.legend(bbox_to_anchor=(1.05, 2.05), loc='lower left', borderaxespad = 0.)
    fig.suptitle('Decision Tree Regressor Learning Performances', fontsize = 16, y = 1.03)
    fig.tight_layout()
    fig.show()
```

* The following code cell produces four graphs for a **Decision tree Regressor** with different maximum of depths. 

* Each graph visualizes the learning curves of the model for both training and testing as the size of the training set is increased. 

* Note that the shaded region of a learning curve denotes the uncertainty of that curve (measured as the standard deviation). The model is scored on both the training and testing sets using R<sup>2</sup>, the coefficient of determination.

```python
# Produce learning curves for varying training set sizes and maximum depths
ModelLearning(features, prices)
```

### Complexity Curves
The following two code cells produce a graph for a decision tree model that has been trained and validated on the training data using different maximum depths. The graph produces two complexity curves â€” one for training and one for validation. Similar to the **learning curves**, the shaded regions of both the complexity curves denote the uncertainty in those curves, and the model is scored on both the training and validation sets using the `performance_metric` function.  

```python
def ModelComplexity(X, y):
    """ Calculates the performance of the model as model complexity increases.
        The learning and testing errors rates are then plotted. """
    
    # Create 10 cross-validation sets for training and testing
    cv = ShuffleSplit(X.shape[0], n_iter = 10,test_size = 0.2,  random_state = 337)

    # Vary the max_depth parameter from 1 to 10
    max_depth = np.arange(1,11)

    # Calculate the training and testing scores
    train_scores, test_scores = curves.validation_curve(DecisionTreeRegressor(), X, y, \
        param_name = "max_depth", param_range = max_depth, cv = cv, scoring = 'r2')

    # Find the mean and standard deviation for smoothing
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    # Plot the validation curve
    plt.figure(figsize=(7, 5))
    plt.title('Decision Tree Regressor Complexity Performance')
    plt.plot(max_depth, train_mean, 'o-', color = 'r', label = 'Training Score')
    plt.plot(max_depth, test_mean, 'o-', color = 'g', label = 'Validation Score')
    plt.fill_between(max_depth, train_mean - train_std, \
        train_mean + train_std, alpha = 0.15, color = 'r')
    plt.fill_between(max_depth, test_mean - test_std, \
        test_mean + test_std, alpha = 0.15, color = 'g')
    
    # Visual esthetics
    plt.legend(loc = 'lower right')
    plt.xlabel('Maximum Depth')
    plt.ylabel('Score')
    plt.ylim([-0.05,1.05])
    plt.show()
```

```python
ModelComplexity(X_train, y_train)
```

### Bias-Variance Tradeoff

We can say that the model is over-simplified and is not capturing the underlying relationships present in the data for both training and validation datasets. 

- Model trained with max_depth of 10: The model suffers from high variance at this depth, and is overfitting on the training data. In the graph, we can see that the training score at this depth is almost equal to 1.0, while the validation score is lower, at around 0.7. The curves also seem to be diverging away from each other at this point.

## Grid Search:

The grid search technique is a systematic way of going through different combinations of parameter values while cross validating the results to determine the parameter combination which gives the best performance based on a error or scoring metrics.

In order to optimize a learning algorithm, we can apply grid search by specifying the parameters, and the possible values of those parameters. The grid search then returns the best parameter values for our model, after fitting the supplied data. This takes out the guess-work involved in seeking out the optimal paramter values for a classifier.

* **Advantages**: It saves time if the total number of values we want to check is **not big**. : 
* **Disadvantages**: **The curse of dimensionality** it repeats the train-validation process each possible Combination. The number of combinations grows exponentially for each new Hyper parameter added to the gird.

* $N:$ The number of parameters we want to include in the grid search 


* $Card(param_i)$: The number of values mentioned in grid search dictionary of the  $param_i$


* $ Num_{possibilities} =$ $\prod\limits_{i=1}^{i=N} Card (param_i)$

## Cross-Validation:

The k-fold crossvalidation training technique is a way of splitting the dataset into *k* partitions of equal size, and then running *k* separate learning experiments on the training data. In each of the experiments, we chose a training set of the size of k-1 partitions, we train our model on these partitions, and evaluate the results on the remaining validation data. The results/scores for the k experiments are then averaged out. 

This technique is benefitial when using grid search to optimize a model because it allows us to look for parameter settings that perform well for different validation sets. If we had a single validation set, it's easy to tune a model to perform well for that specific  set (and result in overfitting on the validation set in this case), while cross validation allows us to generalize the results. It allows also to use the entire data for both training and validation(testing) operations in case the dataset is not splitted earlier into training and testing sets.

We shuffle and split randomly the  $Full_{data}$  into two data sets eg : 80% for $Training_{data}$ and 20% for $Testing_{data}$.  

**The K-fold cross validation begins **:

**$The$ $process$**:

We Shuffle and split the $Training_{data}$ into k equal parts($K$-$fold$).Each time we pick a new $fold_i$ from the $k$-$folds$ not already used for validation. We create a new $model_i$ .We train the $model_i$ on the $(k-1)$ parts that are different from $part_i$.We validate the $model_i$ on that $fold_i$ we picked up earlier.


By repeating **$The$ $process$** mentioned before K times we end up by having K-different models. Each  $model_i$ is trained on the (k-1) $fold_j$ ;$j$ different from $i$;  of the **Training data** and validated just once on that $fold_i$ choosed earlier.By calculating the average of these k-models we end up by a **One final model** that used all the $Training_{data}$ for both learning and validating.

** End K-fold cross validation**

We test the **One final model** on the $Testing_{data}$ to be sur  it doesn't suffer from: **High Bias** or **High Variance**.

* ** The benefits does K-fold cross validation method provide for grid search when optimizing a model :** 
 by using K-fold combined with cross validation method each possible case for hyper-parameters will trained and validated k-times so by the end of the day we will have $k$ $*$ $num_{possibilities}$ model the k-fold cross validation method will choose the best $k$ models and averages them to construct the **one final model**. Form this process we can say that the final model will fit the better the data than using train_test_split combined with grid searchcv the risk that our model will suffer from high variance is very high . 
 
 The **one final model** will have a reduced variance because it uses all the $training_{data}$ for validation. 

### Fitting a Model

Now we'll bring everything together and train a model using the **decision tree regressor algorithm**. To ensure that we are producing an optimized model, we will train the model using the grid search technique to optimize the `'max_depth'` parameter for the decision tree. The `'max_depth'` parameter can be thought of as how many questions the decision tree algorithm is allowed to ask about the data before making a prediction.

```python
# Import 'make_scorer' and 'GridSearchCV'
from sklearn.grid_search import GridSearchCV
from sklearn. metrics import make_scorer

def fit_model(X, y):
    """ Performs grid search over the 'max_depth' parameter for a 
        decision tree regressor trained on the input data [X, y]. """
    
    # Create cross-validation sets from the training data
    # sklearn version 0.18: ShuffleSplit(n_splits=10, test_size=0.1, train_size=None, random_state=None)
    # sklearn versiin 0.17: ShuffleSplit(n, n_iter=10, test_size=0.1, train_size=None, random_state=None)
    cv_sets = ShuffleSplit(X.shape[0], n_iter = 10, test_size = 0.20, random_state = 337)

    # Create a decision tree regressor object
    regressor = DecisionTreeRegressor()

    # Create a dictionary for the parameter 'max_depth' with a range from 1 to 10
    
    params = {'max_depth':range(1,11)}
    

    # Transform 'performance_metric' into a scoring function using 'make_scorer' 
    scoring_fnc = make_scorer(performance_metric)

    # Create the grid search cv object --> GridSearchCV()
    # Make sure to include the right parameters in the object:
    # (estimator, param_grid, scoring, cv) which have values 'regressor', 'params', 'scoring_fnc', and 'cv_sets' respectively.
    grid = GridSearchCV(regressor, params, scoring=scoring_fnc, cv=cv_sets)

    # Fit the grid search object to the data to compute the optimal model
    grid = grid.fit(X, y)

    # Return the optimal model after fitting the data
    return grid.best_estimator_
```

### Making Predictions
Once a model has been trained on a given set of data, it can be used to make predictions on new sets of input data. In the case of a *decision tree regressor*, the model has learned *what the best questions to ask about the input data are*, and can respond with a prediction for the **target variable**.

```python
# Fit the training data to the model using grid search
reg = fit_model(X_train, y_train)

# Produce the value for 'max_depth'
print "Parameter 'max_depth' is {} for the optimal model.".format(reg.get_params()['max_depth'])
```

### Sensitivity
An optimal model is not necessarily a robust model. Sometimes, a model is either too complex or too simple to sufficiently generalize to new data. Sometimes, a model could use a learning algorithm that is not appropriate for the structure of the data given. Other times, the data itself could be too noisy or contain too few samples to allow a model to adequately capture the target variable â€” i.e., the model is underfitted.
